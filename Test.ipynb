```python
import tkinter as tk
from tkinter import ttk

window = tk.Tk()

window.title("Name")

tabControl = ttk.Notebook(window)

tab1 = tk.Frame(tabControl)
tab2 = tk.Frame(tabControl)
tab3 = tk.Frame(tabControl)
tab4 = tk.Frame(tabControl)
tab5 = tk.Frame(tabControl)

# Данні
sub_tab1 = tk.Frame(tab1)
sub_tab2 = tk.Frame(tab1)
tab1_entry_file_entry = tk.Entry(sub_tab1)
tab1_button_select_file = tk.Button(sub_tab1, text = "Завантажити файл")

tab1_result_head = tk.Text(sub_tab2, bd=2, relief="sunken")
tab1_result_plot = tk.Text(sub_tab2, bd=2, relief="sunken")

sub_tab1.grid(row = 0, column = 0)
tab1_entry_file_entry.grid(row = 0, column = 0)
tab1_button_select_file.grid(row = 1, column = 0)

sub_tab2.grid(row = 0, column = 1)
tab1_result_head.grid(row = 0, column = 0, sticky= "nsew")
tab1_result_plot.grid(row = 2, column = 0, sticky= "nsew")

tab2_list= tk.Text(tab2, bd=2, relief="sunken")
tab2_heatmap_plot = tk.Text(tab2, bd=2, relief="sunken")
tab2_list.grid(row = 0, column = 0, sticky= "nsew")
tab2_heatmap_plot.grid(row = 0, column = 1, sticky= "nsew")


tab3.columnconfigure(0, weight=1)  
tab3.columnconfigure(1, weight=3)
tab3.rowconfigure(0, weight=1)  
tab3.rowconfigure(1, weight=3)
var_lr = tk.BooleanVar()
var_rf = tk.BooleanVar()
var_xgb = tk.BooleanVar()
sub_tab3 = tk.Frame(tab3)
tab3_best_model = tk.Text(tab3, bd=2, relief="sunken")
tab3_graphs = tk.Text(tab3, bd=2, relief="sunken")
check_lr = tk.Checkbutton(sub_tab3, text="Linear Regression", variable=var_lr)
check_rf = tk.Checkbutton(sub_tab3, text="Random Forest", variable=var_rf)
check_xgb = tk.Checkbutton(sub_tab3, text="XGBoost", variable=var_xgb)
sub_tab3.grid(row = 0, column = 0, sticky = "w")
tab3_best_model.grid(row = 1, column = 0, sticky= "nsew")
tab3_graphs.grid(row = 0, column = 1, rowspan=2, sticky= "nsew")
check_lr.pack(anchor=tk.W)
check_rf.pack(anchor=tk.W)
check_xgb.pack(anchor=tk.W)

tab4_weights = tk.Text(tab4)
tab4_table_score = tk.Text(tab4)
tab4_ploted_score = tk.Text(tab4)

tab4_weights.grid(row = 0, column = 0, rowspan=2, sticky= "nsew")
tab4_table_score.grid(row = 0, column = 1)
tab4_ploted_score.grid(row = 1, column = 1)

tab5_proposed = tk.Text(tab5)
tab5_rolling = tk.Text(tab5)

tab5_proposed.grid(row = 0, column = 0)
tab5_rolling.grid(row = 0, column = 1)

tabControl.add(tab1, text = 'Дані')
tabControl.add(tab2, text = 'Кореляція')
tabControl.add(tab3, text = 'Моделі')
tabControl.add(tab4, text = 'Скоринг')
tabControl.add(tab5, text = 'Кластеризація')

tabControl.pack(expand=1, fill="both")

window.mainloop()
```
